1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

# Визначаємо символічну змінну
x_sym = sp.symbols('x')

# Лінійна функція
f_lin = 10 * x_sym -5
derivative_f_lin = sp.diff(f_lin,x_sym)

print("похідна:", derivative_f_lin)

x = np.linspace(0,10,100)
plt.plot(x, 10*x - 5)
plt.plot(x, np.full_like(x, 10));
```

за графіком функції можна зробити висновок, що зі збільшенням кількости годин, збільшується і оцінка

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x_sym = sp.symbols('x')

# синус
f_sin = sp.sin(x_sym)
derivative_f_sin = sp.diff(f_sin, x_sym)

# натуральний логарифм
f_ln = sp.log(x_sym)
derivative_f_ln = sp.diff(f_ln, x_sym)


f_sin_num = sp.lambdify(x_sym, f_sin, 'numpy')
d_sin_num = sp.lambdify(x_sym, derivative_f_sin, 'numpy')

f_ln_num = sp.lambdify(x_sym, f_ln, 'numpy')
d_ln_num = sp.lambdify(x_sym, derivative_f_ln, 'numpy')


# Графік sin(x)
plt.figure(figsize=(10, 4))
plt.plot(x, f_sin_num(x), label='sin(x)')
plt.plot(x, d_sin_num(x), label="cos(x)")
plt.legend()
plt.grid(True)
plt.title('sin(x) та його похідна')
plt.show()

# Графік ln(x)
plt.figure(figsize=(10, 4))
plt.plot(x, f_ln_num(x), label='ln(x)')
plt.plot(x, d_ln_num(x), label='1/x')
plt.legend()
plt.grid(True)
plt.title('ln(x) та його похідна')
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')

# Функція та похідна
f_x = x**2 - 5*x + 1
derivative_f_x = sp.diff(f_x, x)

print("похідна:", derivative_f_x)

f_num = sp.lambdify(x, f_x, 'numpy')
d_num = sp.lambdify(x, derivative_f_x, 'numpy')

# Числові значення x
x = np.linspace(0, 100)

plt.plot(x, f_num(x), label='f(x)=x²-5x+1')
plt.plot(x, d_num(x), label="f'(x)=2x-5")

plt.legend()
plt.grid(True)
plt.show()
```
